```python
# User Login Check
#Applications:Banking apps authentication

username = "admin"
password = "1234"

# Using nested if instead of AND
if username == "admin":
    if password == "1234":
        print("Login Successful")
    else:
        print("Invalid Credentials")
else:
    print("Invalid Credentials")
```

```python
# Pass / Fail Analyzer
#Applications:Online exam systems

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for i in range(len(marks)):
    if marks[i] >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

print("Total Students:", len(marks))
print("Number of Pass Students:", pass_count)
print("Number of Fail Students:", fail_count)
```

```python
# Simple Data Cleaner
#Applications:Removing formatting issues in Excel sheets

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    name = name.strip()       
    name = name.lower()        
    cleaned_names.append(name)

print("Cleaned List:", cleaned_names)
```

```python
# Message Length Analyzer
#Applications:Chat moderation systems

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)
    
    if length > 10:
        print("This message is longer than 10 characters")
    
    print("----------------------")
```

```python
# Error Message Detector
#Applications:Software debugging logs

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR Count:", error_count)
```
