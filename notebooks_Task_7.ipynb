```python
##Smart Light Controller

class SmartLight:
    def __init__(self, light_name, status):
        self.light_name = light_name
        self.status = status

    def display(self):
        print(f"{self.light_name} is {self.status}")

# main
light = SmartLight("Bedroom Light", "ON")
light.display()
```

```python
##Employee ID Card System
class EmployeeID:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


emp = EmployeeID("Rahul", "EMP102", "IT")
emp.display()
```

```python
##Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c = Contact("Anita", "9876543210")
c.display()
```

```python
##Product Price Tag Generator

class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def display(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))


p = Product("Headphones", 2499)
p.display()
```

```python
##Movie Rating Display System

class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


m = Movie("Inception", 4.8)
m.display()
```

```python
##Delivery Address Manager

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


d = Delivery("Suman", "Hyderabad")
d.display()
```
