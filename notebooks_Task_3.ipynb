```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Highest score
max_score = max(employees.values())

# Employees with highest score
top_performers = [name for name, score in employees.items() if score == max_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase and remove punctuation
query = query.lower()
query = query.translate(str.maketrans('', '', string.punctuation))

words = query.split()

# Count frequency
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

# Display words searched more than once
result = {k: v for k, v in freq.items() if v > 1}

print(result)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentage
total = len(emails)
for domain, count in domain_count.items():
    percent = (count / total) * 100
    print(f"{domain}: {percent:.0f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales = sum(sales) / len(sales)

threshold = avg_sales * 1.3

for i, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {i}: {value}")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# Count IDs
for uid in user_ids:
    count[uid] = count.get(uid, 0) + 1

# Show duplicates
for uid, c in count.items():
    if c > 1:
        print(f"{uid} → {c} times")
```
