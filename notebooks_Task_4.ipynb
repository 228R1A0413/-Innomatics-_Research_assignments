```python
##Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False

    if amount in valid_plans:
        return True

    return False


while True:
    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        print("Recharge Successful ")
        break
    else:
        print("Invalid recharge. Try again ")
```

```python
##Inventory Reorder Alert System
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, "→ Reorder Alert ")
        else:
            print(item, "→ Stock OK ")


products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 30
}

check_inventory(products)
```

```python
##Student Result Processing System

def student_result(marks):
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [60, 55, 40, 70]

result = student_result(marks)

print("Result:", result)
```

```python
##Cab Fare Estimator with Retry Option

def calculate_fare(km, peak):
    fare = 50 + (km * 12)

    if peak.lower() == "yes":
        fare = fare * 1.25

    return fare


while True:
    km = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ")

    fare = calculate_fare(km, peak)

    print("Estimated Fare: ₹", fare)

    retry = input("Retry? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
##Employee Attendance Eligibility Checker

def attendance_check(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percent = (present / len(attendance)) * 100

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

print(attendance_check(attendance))
```

```python
##Password Strength Checker

def password_strength(password):
    special_chars = "@#$"

    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


pwd = input("Enter password: ")
print(password_strength(pwd))
```
